```python
def removepunctuation(text):
  punctuation_free=""
  for i in text:
    if i not in string.punctuation:
      punctuation_free=punctuation_free+i
  
  return punctuation_free.lower()
```

```python
def removenumbers(text):
  output=""
  for i in text:
    if not i.isdigit():
      output=output+i
  
  return output
```
